# Procesamiento de señal de voz

```python
%matplotlib inline
from scipy.io.wavfile import read
from pylab import specgram
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = read('grab.wav')
```

```python
plt.plot(A[1])
plt.show()
```

```python
espectro = specgram(A[1])
```

```python
def split(arr, size):
     arrs = []
     while len(arr) > size:
         pice = arr[:size]
         arrs.append(pice)
         arr   = arr[size:]
     arrs.append(arr)
     return arrs
    
Phones = split(A[1], 100)
```

```python
plt.plot(Phones[1])
plt.show()
```

```python
fono1 = specgram(Phones[1])
```
