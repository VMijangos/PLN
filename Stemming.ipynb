# Stemming

```python
from nltk.stem.snowball import SnowballStemmer

stemizer = SnowballStemmer('spanish')
```

```python
text = open('texto.txt','r').read().decode('utf8')
print text
```

```python
text = text.replace('.','')
text = text.replace(',','')
text = text.replace('(','')
text = text.replace(')','')
text = text.replace(';','')
text = text.replace('-',' ')

print text
```

```python
tokens = text.split()

for token in tokens:
    print token
```

```python
for token in tokens:
    print stemizer.stem(token)
```

```python
from collections import defaultdict

Stems = defaultdict(list)
for token in set(tokens):
    Stems[stemizer.stem(token)].append(token)
    
for stem, values in Stems.iteritems():
    print stem
    for token in values:
        print '\t', token
```
